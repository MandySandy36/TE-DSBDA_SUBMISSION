## Import Required Library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Read Titanic dataset into Dataframe

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.shape
```

```python
titanic.head()
```

## Set the style for the plots

```python
sns.set_style("whitegrid")
```

## 1. Survival countplot

```python
plt.figure(figsize=(6,4))
sns.countplot(data=titanic, x="survived", hue="survived", palette="coolwarm", legend=False)
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
# This tells you how many people survived vs. died.
# If more died (which is expected), the "0" bar will be taller or vice-versa.
# It gives a quick overview of the tragedy’s scale.
```

## 2. Survival by Gender

```python
plt.figure(figsize=(6,4))
sns.countplot(x="sex", hue="survived", data=titanic, palette="coolwarm")
plt.title("Survival by Gender")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```

```python
# Look at survival difference between males and females.
# In Titanic, females had higher survival rates due to "women and children first."
# If one gender has a much higher "Yes" bar, that’s a survival bias.
```

## 3. Survival by Passenger Class

```python
plt.figure(figsize=(6,4))
sns.countplot(x="pclass", hue="survived", data=titanic, palette="coolwarm")
plt.title("Survival by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```

```python
# See if wealth/class affected survival.
# 1st class likely has more survivors.
# If 3rd class has more deaths, this may reflect social inequality at the time.
```

## 4. Age Distribution of Passengers

```python
plt.figure(figsize=(8,5))
sns.histplot(titanic["age"].dropna(), kde=True, bins=30, color="purple")
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Density")
plt.show()
```

```python
# Spot age groups with higher survival.
# Look for younger passengers, especially children, surviving more.
# Also helps you decide if you want to create age groups (like "child", "adult", etc.)
```

## 5. Survival based on Fare Paid

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=titanic, x="survived", y="fare", hue="survived", palette="coolwarm", legend=False)
plt.title("Fare vs. Survival")
plt.xticks([0, 1], ["Did Not Survive", "Survived"])
plt.show()
```

```python
# Boxplots show fare distribution — min, max, median, and outliers.
# If survivors paid more on average, you’ll see higher fare values for survived = 1.
# Tells you if money increased chances of survival (linked to class too).
```
