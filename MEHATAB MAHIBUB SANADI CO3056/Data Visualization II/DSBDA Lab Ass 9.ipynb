## Import Required Library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Read Titanic dataset into Dataframe

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic = titanic[['sex','age','survived']]
```

```python
titanic.head()
```

## Set the style for the plots

```python
sns.set_style("whitegrid")
```

## 1. Box Plot: Age Distribution by Gender and Survival

```python
plt.title("Age Distribution by Gender")
sns.boxplot(x='sex',y='age',data=titanic)
plt.xlabel("Gender")
plt.ylabel("Age")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```

```python
plt.title("Age Distribution by Gender and Survival")
sns.boxplot(data=titanic, x="sex", y="age", hue="survived", palette="Set2")
plt.xlabel("Gender")
plt.ylabel("Age")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```

## 2. Observations and Inference

Female Passengers
1. Survivors (green box) tend to have a wide age range, but there is a visible cluster of survivors in the young to middle-age range (teens to 40s).
2. The non-survivors group (pink box) is smaller, meaning more women survived.

Male Passengers
1. Non-survivors dominate (larger pink box), showing that more men died overall.
2. Among male survivors, the age range is more concentrated around 2040, though there are survivors from both young and older age groups.

General Inference
1. Gender had a significant impact on survival: women had much higher survival rates.
2. Children and young adults of both genders had relatively better survival chances.
3. This aligns with the historical "women and children first" evacuation policy.
