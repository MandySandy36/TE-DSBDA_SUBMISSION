## Import Required Library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Read Iris dataset into Dataframe

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

## 1. Features and Their Types

```python
iris.info()
```

## Set the style for the plots

```python
sns.set_style("whitegrid")
```

## 2. Histograms for Each Feature

```python
plt.suptitle("Histograms of Iris Features", fontsize=16)
iris.hist(figsize=(10, 6), color='skyblue', edgecolor='black', bins=15)
plt.tight_layout()
plt.show()
```

## 3. Boxplots for Each Feature

```python
# One boxplot per feature
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

plt.figure(figsize=(12, 8))
for i, col in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(y=iris[col], color='lightgreen')
    plt.title(f'Boxplot of {col}')

plt.tight_layout()
plt.show()
```

## 4. Compare distributions and identify outliers.

1. sepal_width shows clear outliers (a few very low or high values).
2. petal_length and petal_width are tightly packed for certain classes (like setosa).
3. The boxplots show median lines, spread (IQR), and potential outliers (dots outside whiskers).
